```python
import pandas as pd
df = pd.read_csv('https://api.kite.trade/instruments').drop(columns=['tick_size', 'last_price'])
df=df[['name','segment','tradingsymbol','lot_size','expiry','strike','instrument_type','exchange_token','instrument_token','exchange']]
df = df[df['segment'].isin(["BFO-FUT","NFO-FUT","MCX-FUT","BFO-OPT", "NFO-OPT", "MCX-OPT"])]
df['strike']=df['strike'].astype(int)

pairs = [
#future
("NIFTY","NFO-FUT"),("BANKNIFTY","NFO-FUT"),("FINNIFTY","NFO-FUT"),("MIDCPNIFTY","NFO-FUT"),("NIFTYNXT50","NFO-FUT"),("SENSEX","BFO-FUT"),("BANKEX","BFO-FUT"),("CRUDEOIL","MCX-FUT")
#option
,("NIFTY","NFO-OPT"),("BANKNIFTY","NFO-OPT"),("FINNIFTY","NFO-OPT"),("MIDCPNIFTY","NFO-OPT"),("NIFTYNXT50","NFO-OPT"),("SENSEX","BFO-OPT"),("BANKEX","BFO-OPT"),("CRUDEOIL","MCX-OPT")

]


trade = pd.concat([df[(df["name"]==n) & (df["segment"]==s)].head(1) for n,s in pairs], ignore_index=True)

# Filter for rows where name is 'CRUDEOIL'
crudeoil_df = trade[trade['name'] == 'CRUDEOIL']

print(crudeoil_df,"\n\n\n\n",trade)
```
